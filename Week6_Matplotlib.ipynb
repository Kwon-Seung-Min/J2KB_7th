- ÏòàÏãúÎ°ú **k-.v**Îäî Í≤ÄÏ†ïÏÉâ ÎåÄÏâ¨Îã∑ ÎùºÏù∏ Ïä§ÌÉÄÏùºÏùÑ Í∞ÄÏßÑ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Í≥†, ÌïµÏã¨ ÏõêÏÜåÏóê Ïó≠ÏÇºÍ∞ÅÌòïÏùò ÎßàÏª§Î•º Ï∞çÏñ¥ ÌëúÏãúÌïòÎèÑÎ°ù ÌïúÎã§Îäî ÏùòÎØ∏Ïù¥Îã§.

## MatplotlibÏóêÏÑú ÌïúÍ∏ÄÏùÑ ÌëúÏãúÌïòÎäî Î∞©Î≤ï

MatplotlibÏóêÏÑú ÌïúÍ∏ÄÏùÑ ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî, ÏÇ¨Ïö©Ìï† Ìè∞Ìä∏Î•º ÎØ∏Î¶¨ ÏßÄÏ†ïÌï¥Ï£ºÏñ¥Ïïº ÌïúÎã§.

```python
# ÎßëÏùÄÍ≥†Îîï Í∏ÄÍº¥ÏùÑ Ï†ÅÏö©
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot([1, 1.25, 1.5, 1.75, 2], [1, 1.5, 2, 2.5, 3], "k-.v")
plt.plot([1, 3, 6, 10, 15], [1, 2, 3, 4, 5], "h:b")
plt.xlabel("X Ï∂ï")
plt.ylabel("Y Ï∂ï")
plt.title("ÌïòÎÇòÏùò ÏòÅÏó≠Ïóê Îëê Í∞úÏùò Í∑∏ÎûòÌîÑ ÎèôÏãúÏóê Í∑∏Î¶¨Í∏∞")
plt.legend(["A Í∑∏ÎûòÌîÑ", "B Í∑∏ÎûòÌîÑ"])
plt.show()
```

## MatplotlibÏóêÏÑú ÎßâÎåÄÍ∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
def barplot():
    plt.bar([1, 2, 3, 4, 5], [10, 15, 25, 40, 60], color="green", width=0.5, antialiased=True, fill=True)
    plt.xlabel("X Ï∂ï")
    plt.ylabel("Y Ï∂ï")
    plt.title("ÎßâÎåÄÍ∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞")
    plt.legend(["A Í∑∏ÎûòÌîÑ", "B Í∑∏ÎûòÌîÑ"])
    plt.show()

barplot()
```

## MatplotlibÏóêÏÑú ÏõêÌòïÍ∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
def piegraph():
    plt.pie([10, 20, 30, 40], labels=['A', 'B', 'C', 'D'], shadow=False, colors=['red', 'yellow', 'green', 'magenta'], autopct='%.2f%%')
    plt.xlabel("X Ï∂ï")
    plt.ylabel("Y Ï∂ï")
    plt.title("ÏõêÌòïÍ∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞")
    plt.legend(["A Í∑∏ÎûòÌîÑ", "B Í∑∏ÎûòÌîÑ", "C Í∑∏ÎûòÌîÑ", "D Í∑∏ÎûòÌîÑ"])
    plt.show()

piegraph()
```
